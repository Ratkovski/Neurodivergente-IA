```python
import google.generativeai as genai
from google.colab import userdata
api_key = userdata.get('SECRET_KEY')
genai.configure(api_key=api_key)
```

```python
#Listando os modelos dispon√≠veis
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(m.name)
```

```python
generation_config={
    "temperature": 0.5,
    "candidate_count":1,
}
```

```python
safety_settings={
    'HATE':'BLOCK_NONE',
    'HARASSMENT':'BLOCK_NONE',
    'SEXUAL':'BLOCK_NONE',
    'DANGEROUS':'BLOCK_NONE',

    }
```

```python
system_instruction ="Voc√™ √© um especialista em neurodiverg√™ncia, com especializa√ß√µes em neuro-aprendizagem, phd em psicologia educacional e comportamental o qual vai criar as atividades que forem solicitadas de acordo com as informa√ß√µes coletadas "
model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest",
                               generation_config=generation_config,
                              system_instruction=system_instruction,
                               safety_settings=safety_settings)

```

```python
# pegando dados do aluno
# Para o futuro : pegar mais caracteristicas que sejam importantes para ajudar na organiza√ß√£o do plano de estudo
def get_informacao_usuario():
  neurodivergencia = input("Qual o tipo de neurodiverg√™ncia (Exemplo: autismo, TDAH, dislexia)? ")
  idade = int(input("Qual a idade do aluno? "))
  assunto = input("Qual o assunto a ser ensinado? ")
  nivel = input("Nivel de dificuldade de estudo ")
  return neurodivergencia,idade,assunto,nivel
```

```python
def processar_resposta(resposta):
    print(resposta.text)
```

```python
def main():
    chat = model.start_chat(history=[])
    #verifica se a pessoa que continuar a pesquisa
    prosseguir = input("Gostaria de prosseguir? Digite 'yes': ")
    if prosseguir.lower() == "yes":
        while True:
            neurodivergencia, idade, assunto, nivel = get_informacao_usuario()
        #monta o prompt para podermos ter a resposta do model
            prompt = (
                f"Gostaria de um planejamento de estudo para uma pessoa que tem a "
                f"neurodiverg√™ncia {neurodivergencia} e sua idade √© {idade} a qual "
                f"gostaria de aprender sobre {assunto} e o n√≠vel de dificuldade seja {nivel}. "
                f"Inclua links do YouTube: "
            )
        #envia o prompt e monta a resposta
            response = chat.send_message(prompt)
            processar_resposta(response)

            #verifica se a pessoa que continuar a pesquisa ou quer sair do sistema
            continuar = input("Gostaria de continuar? Digite 'yes' para continuar ou qualquer outra coisa para sair: ")
            if continuar.lower() != "yes":
                break
    else:
        print("Saindo do sistema.")

if __name__ == "__main__":
    main()
```
